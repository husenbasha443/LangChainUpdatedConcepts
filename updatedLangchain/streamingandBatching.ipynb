```python
from dotenv import load_dotenv
load_dotenv()

from langchain_groq import ChatGroq

model = ChatGroq(model="llama-3.1-8b-instant")
```

```python
model.invoke("Explain Machine Learning about 500 words")
```

```python
for chunk in model.stream("Explain Machine Learning about 500 words"):
    print(chunk.text)
```
