```python
from dotenv import load_dotenv
load_dotenv()

from langchain.chat_models import init_chat_model

model = init_chat_model(
    model="llama-3.1-8b-instant",
    model_provider="groq",
)
```

##### Perform Pydantic

```python
from pydantic import BaseModel,Field

class Movie(BaseModel):
    name: str=Field(description="The name of the movie")
    year: int=Field(description="The year of the movie")
    director: str=Field(description="The director of the movie")
    rating: float=Field(description="The rating of the movie")
```

```python
structured_output = model.with_structured_output(Movie) 
```

```python
result = structured_output.invoke("Provide the name, year, director, and rating of a movie KGF.")
result
```
