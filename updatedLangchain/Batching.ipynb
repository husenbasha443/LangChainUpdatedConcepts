```python
from dotenv import load_dotenv
load_dotenv()

from langchain_groq import ChatGroq

model = ChatGroq(model_name="llama-3.1-8b-instant")
```

#### Batch without Config

```python
model.batch([
    "What is JAVA",
    "What is Python",
    "What is C++"
])
```

#### Batch With Config

```python
model.batch([
    "What is JAVA",
    "What is Python",
    "What is C++"
    "What is C",
    "What is C#",
    "What is JavaScript",
    "What is TypeScript",],

    config={
        'max_concurrency': 2
    })
    
```
